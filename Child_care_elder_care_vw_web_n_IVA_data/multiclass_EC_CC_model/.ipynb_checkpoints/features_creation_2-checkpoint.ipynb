```python
# Iterate over the dictionary items
for key, phrases in data.items():
    # col_name = 'web_' + key
    col_name = 'iva_' + key

    
    # Replace non-string values with an empty string
    # df['page_name'] = df['page_name'].apply(lambda x: '' if pd.isna(x) or isinstance(x, float) else x)
    df['input'] = df['input'].apply(lambda x: '' if pd.isna(x) or isinstance(x, float) else x)

    
    # Apply the condition to create the new column
    # df[col_name] = df['page_name'].apply(lambda text: 1 if any(phrase in text.lower() for phrase in phrases) else 0)
    df[col_name] = df['input'].apply(lambda text: 1 if any(phrase in text.lower() for phrase in phrases) else 0)


# Output the updated DataFrame
df.head()
```
